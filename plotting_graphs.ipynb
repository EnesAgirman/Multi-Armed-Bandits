```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
# take the integral of the function f between 0 and 1
k = 3
l = 11
f = ( (p**k) * (1-p)**l ) * math.comb(k+l, k)

# integrate f from 0 to 1
integral = np.trapz(f, p)
print(f'integral of f from 0 to 1: {integral}')
```

```python
# plot p^k (1-p)^l in range 0-1
p = np.linspace(0, 1, 1000)
k = 3
l =11
y = p**k * (1-p)**l
y = y * math.comb(k+l, k)
# take the integral of the function f between 0 and 1
k = 3
l = 11
f = ( (p**k) * (1-p)**l ) * math.comb(k+l, k)
# integrate f from 0 to 1
integral = np.trapz(f, p)

# make the integral of f from 0 to 1 equal to 1
y = y / integral

# plot the graph
plt.plot(p, y)
plt.xlabel('p')
plt.ylabel(f'normalized p^{k} (1-p)^{l}')
plt.title(f'normalized p^{k} (1-p)^{l} graph')
plt.show() 

integral = np.trapz(y, p)
print(f'integral of y from 0 to 1: {integral}')
```

```python
# draw a very noisy function from 0 to 1. Just random noise
x = np.linspace(0, 1, 100)
y = np.random.rand(100)
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('reward')
plt.title('Very Noisy Function')
```
